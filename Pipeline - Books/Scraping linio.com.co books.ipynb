# Extract

```python
import bs4
import requests
import pandas as pd
```

```python
responses = []
for i in range(1, 18):
    response = requests.get("https://www.linio.com.co/c/libros/literatura-y-novelas?page={}".format(i))
    responses.append(response)
```

```python
soups = []
for i in range(len(responses)):
    soup = bs4.BeautifulSoup(responses[i].text, "html.parser")
    soups.append(soup)
```

```python
titles_book = []
for i in range(len(soups)):
    titles = soups[i].select("span.title-section")
    for title in titles:
        titles_book.append(title.string)
```

```python
titles_book
```

```python
prices_book = []
for i in range(len(soups)):
    prices = soups[i].select("span.price-main")
    for price_1 in prices:
        prices_book.append(price_1.text)
```

```python
prices_book[0:6]
```

```python
len(prices_book)
```

# Transform

```python
idx = []
for i in range(1, len(prices_book)+1):
    idx.append(i)
```

```python
df = pd.DataFrame({"name":titles_book, "price":prices_book, "idx":idx})
```

```python
df[0:5]
```

```python
df["price"] = df["price"].apply(lambda name: name.split("\n")[0])
```

```python
df[0:5]
```

```python
df["price"] = (df["price"].apply(lambda name: name.split("$")[1])
               .map(lambda point:point.split("."))
               .map(lambda unit:''.join(unit)).astype(int))
```
